```python
def dataLoad(txt, file, inps, name):
    ct = datetime.datetime.now()
    print(str(ct)+';'+txt+" Executing DataLoad(",inps,")")
    f.write(str(ct)+';'+txt+" Executing DataLoad("+inps+")\n")

    
    DataTable=pd.read_csv(file)
    NoOfDefects=DataTable.shape[0]
    
    store[name +'.DataTable'] = DataTable
    store[name +'.NoOfDefects'] = NoOfDefects
    
#     print(name+'.DataTable')
                   
    
    ct = datetime.datetime.now()
    print(str(ct)+';'+txt+" Exit")
    f.write(str(ct)+';'+txt+" Exit\n")
    
def timeFunction(txt, val, inps):
    ct = datetime.datetime.now()
    print(str(ct)+';'+txt+" Executing TimeFunction(",inps,")")
    f.write(str(ct)+';'+txt+" Executing TimeFunction("+inps+")\n")
    
    time.sleep(int(val))
    
    ct = datetime.datetime.now()
    print(str(ct)+';'+txt+" Exit")
    f.write(str(ct)+';'+txt+" Exit\n")
```

```python
def fun(name, obj, stack):
    
#     print(name, obj)
    
    if(obj['Type']=='Task'):

        str_ = ''
        for it in stack:
            str_ += it+'.'  
            
        ct = datetime.datetime.now()
        print(str(ct)+';' + str_ + name + ' Entry') #TASK entry
        f.write(str(ct)+';' + str_ + name + ' Entry\n')
        
        inps=[]
        for ele in obj['Inputs']:
            inps.append(ele)
        
        temp=""
        
        for ele in inps[:-1]:
            temp+=str(obj['Inputs'][ele])+','
        temp+=str(obj['Inputs'][inps[-1]])
        
        if 'Condition' in obj:
            cond = obj['Condition']
            ref = cond.split()
            
            path=ref[0][2:-1]
            op=ref[1]
            val=int(ref[-1])
            
#             print("\n\n\n",store[path])
#             print(val)
            
            if((op == '<' and store[path] < val) or (op=='>' and store[path]>val)):
            
                
                if(obj['Function']=="DataLoad"):
#                     print("\n\n ",name,"\n\n")
                    p = ""
                    for ele in stack:
                        p= p+ele+'.'
                        
                    dataLoad(str_+name,obj['Inputs']['Filename'],temp, p+name)

                elif(obj['Function']=="TimeFunction"):         
                    timeFunction(str_+name,obj['Inputs']['ExecutionTime'],temp)
            else:
                
                f.write(str(datetime.datetime.now())+';'+str_+name+' Skipped\n')
                f.write(str(datetime.datetime.now())+';'+str_+name+' Exit\n')
                
        else:
            if(obj['Function']=="DataLoad"):
                    p = ""
                    for ele in stack:
                        p= p+ ele +'.'
                        
                    dataLoad(str_+name,obj['Inputs']['Filename'],temp, p+name)

            elif(obj['Function']=="TimeFunction"):         
                    timeFunction(str_+name,obj['Inputs']['ExecutionTime'],temp) 
                
            

    if(obj['Type']=='Flow'):
        
        str_ = ""
        for it in stack:
            str_ += it+'.'
            
        ct = datetime.datetime.now()
        print(str(ct)+';' + str_ + name + ' Entry') #FLOW entry        
        f.write(str(ct)+';' + str_ + name + ' Entry\n')
        
        
        
        if(obj['Execution']=="Sequential"):
            for act in obj['Activities']:
                stack_2 = copy.deepcopy(stack)
                stack_2.append(name)
                fun(act, obj['Activities'][act],stack_2)
            
                
        if(obj['Execution']=="Concurrent"):
            
            acts = obj['Activities']
            threads = []
            
            for ele in acts:
                stack_2 = copy.deepcopy(stack)
                stack_2.append(name)
                t = threading.Thread(target=fun, args=(ele, acts[ele], stack_2))
                threads.append(t)
                t.start()
                
            for t in threads:
                t.join()
                

        
        ct = datetime.datetime.now()
        print(str(ct)+';' + str_ + name + ' Exit') #FLOW exit
        f.write(str(ct)+';' + str_ + name + ' Exit\n')
```

```python
f = open("Milestone2A_log.txt","w+")
start = list(data.keys())[0]
fun(start, data[start], [])
f.close()
```
